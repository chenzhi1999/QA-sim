```python
import sys, os
sys.path.insert(0,"/home/zhi/SULI21")
import quasim as qua
import numpy as np

import matplotlib.pyplot as plt
```

```python
qua = qua.QuaTel()
```

```python
lam = 1.0*10**(-6)
s1 = 300.0  # 300 JY
s2 = 200.0  # 200 JY

#[RA,DEC]
pos1 = [[0.23,0.655],[0.21,0.66]]
pos2 = [[0.23,2.280],[0.21,2.28005]]
pos3 = [[0.55,0.665],[0.52,0.66]]

T = 10.0                  #Period of observation[sec]
```

```python
num_pos1,t1 = qua.get_num_photon(pos1,lam,s1,s2,T,'pos')
num_neg1,t1 = qua.get_num_photon(pos1,lam,s1,s2,T,'neg')

num_pos2,t2 = qua.get_num_photon(pos2,lam,s1,s2,T,'pos')
num_neg2,t2 = qua.get_num_photon(pos2,lam,s1,s2,T,'neg')

num_pos3,t3 = qua.get_num_photon(pos3,lam,s1,s2,T,'pos')
num_neg3,t3 = qua.get_num_photon(pos3,lam,s1,s2,T,'neg')
```

```python

plt.figure(figsize=(10,10))
plt.xlim([9.0,9.1])

plt.plot(t1[t1.size/2:],num_pos1[t1.size/2:],label='pos1')
plt.plot(t1[t1.size/2:],num_neg1[t1.size/2:],label='neg1')

plt.plot(t2[t2.size/2:],num_pos2[t2.size/2:],label='pos2')
plt.plot(t2[t2.size/2:],num_neg2[t2.size/2:],label='neg2')

plt.plot(t3[t3.size/2:],num_pos3[t3.size/2:],label='pos3')
plt.plot(t3[t3.size/2:],num_neg3[t3.size/2:],label='neg3')

plt.legend() 
```
