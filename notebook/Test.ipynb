```python
import sys, os
sys.path.insert(0,"/home/zhi/SULI21")
import quasim as qua
import numpy as np
#import pandas as pd

import matplotlib.pyplot as plt
```

```python
QT = qua.QuaTel()
#BSC = qua.BSC_process("/home/zhi/SULI21/bsc5.dat")
#Output as [[star_num1,RA1,DEC1,S1],[star_num2,RA2,DEC2,S2],....] 
#in rad and Jy, S is in V-band
#BSC.pos_s   
```

```python
lam = 1.0*10**(-6)

#[RA,DEC,RADIUS] in rad and meter
pos_t_EW = [[-1.3,0.1,6.4*10**(6)],[-1.30003,0.1,6.39995*10**(6)]]   #SEPARATED BY E-W
pos_t_NS = [[-1.3,0.1,6.4*10**(6)],[-1.3,0.10003,6.40002*10**(6)]]
pos_t_SE = [[-1.3,0.1,6.4*10**(6)],[-1.30002,0.10002,6.399977*10**(6)]]

#NEW YORK COORD but the overall baseline is E-W, with B ~ 270m
pos_t_NY = [[-1.3,0.71,6.0*10**(6)],[-1.300016,0.7099713,6.0002*10**(6)]]  

#[RA,DEC,S] in rad and Jy for flux density

pos1 = [[0.0,0.005,300.0],[0.0,0.01,200.0]]         #EW
pos2 = [[0.0,0.78,300.0],[0.001,0.78,200.0]]      #NS
pos3 = [[-0.52,0.33,300.0],[-0.5,0.331,200.0]]

#Combine all three pairs above
pos = [pos1,pos2,pos3]

T = 10.0  #Period of observation[sec]
```

```python
num_NY,t_NY,Baseline_NY = QT.get_num_photon(pos,pos_t_NY,lam,T,'pos')
num_EW,t_EW,Baseline_EW = QT.get_num_photon(pos,pos_t_EW,lam,T,'pos')
num_NS,t_NS,Baseline_NS = QT.get_num_photon(pos,pos_t_NS,lam,T,'pos')
num_SE,t_SE,Baseline_SE = QT.get_num_photon(pos,pos_t_SE,lam,T,'pos')
```

```python
print("Baseline_EW:", Baseline_EW, "Baseline_NS:", Baseline_NS, 
      "Baseline_SE:", Baseline_SE, "Baseline_NY:", Baseline_NY)
```
