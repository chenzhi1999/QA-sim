```python
import sys, os
sys.path.insert(0,"/home/zhi/SULI21")
import quasim as qua
import numpy as np
#import pandas as pd

import matplotlib.pyplot as plt
```

```python
qua = qua.QuaTel()
```

```python
lam = 1.0*10**(-6)
s1 = 300.0  
s2 = 200.0  

#[RA,DEC,RADIUS] in rad and meter
pos_t_EW = [[-1.3,0.1,6.4*10**(6)],[-1.30003,0.1,6.39995*10**(6)]]   #SEPARATED BY E-W
pos_t_NS = [[-1.3,0.1,6.4*10**(6)],[-1.3,0.10003,6.40002*10**(6)]]
pos_t_SE = [[-1.3,0.1,6.4*10**(6)],[-1.30002,0.10002,6.399977*10**(6)]]

#NEW YORK COORD but the overall baseline is E-W, with B ~ 270m
pos_t_NY = [[-1.3,0.71,6.0*10**(6)],[-1.300016,0.7099713,6.0002*10**(6)]]  

#[RA,DEC] in rad
pos1 = [[0.001,0.52],[0.001,0.525]]
pos2 = [[-0.005,0.52],[-0.01,0.520]]
pos3 = [[1.52,0.73],[1.525,0.735]]


T = 10.0  #Period of observation[sec]
```

```python
num_pos1_NY,t1_NY,Baseline_NY = qua.get_num_photon(pos1,pos_t_NY,lam,s1,s2,T,'pos')
num_pos2_NY,t2_NY,Baseline_NY = qua.get_num_photon(pos2,pos_t_NY,lam,s1,s2,T,'pos')
num_pos3_NY,t3_NY,Baseline_NY = qua.get_num_photon(pos3,pos_t_NY,lam,s1,s2,T,'pos')
print(Baseline_NY)
```

```python

plt.figure(figsize=(10,10))
plt.xlim([1.0,1.2])

plt.plot(t1_NY[:],num_pos1_NY[:],label='pos1, D_DEC')
plt.plot(t2_NY[:],num_pos2_NY[:],label='pos2, D_RA')
plt.plot(t3_NY[:],num_pos3_NY[:],label='pos3')

plt.xlabel("time [sec]");plt.ylabel("Number of Coincidence")
plt.legend() 
```

```python
num_pos1_EW,t1_EW,Baseline_EW = qua.get_num_photon(pos1,pos_t_EW,lam,s1,s2,T,'pos')
num_pos2_EW,t2_EW,Baseline_EW = qua.get_num_photon(pos2,pos_t_EW,lam,s1,s2,T,'pos')

num_pos1_NS,t1_NS,Baseline_NS = qua.get_num_photon(pos1,pos_t_NS,lam,s1,s2,T,'pos')
num_pos2_NS,t2_NS,Baseline_NS = qua.get_num_photon(pos2,pos_t_NS,lam,s1,s2,T,'pos')

num_pos1_SE,t1_SE,Baseline_SE = qua.get_num_photon(pos1,pos_t_SE,lam,s1,s2,T,'pos')
num_pos2_SE,t2_SE,Baseline_SE = qua.get_num_photon(pos2,pos_t_SE,lam,s1,s2,T,'pos')

# [x,y,z]
print("Baseline_EW:", Baseline_EW, "Baseline_NS:", Baseline_NS, "Baseline_SE:", Baseline_SE)
```
