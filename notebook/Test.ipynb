```python
import sys, os
sys.path.insert(0,"/home/zhi/SULI21")
import quasim as qua
import numpy as np

import matplotlib.pyplot as plt
```

```python
qua = qua.QuaTel()
```

```python
lam = 1.0*10**(-6)
s1 = 300.0  
s2 = 200.0  

#[RA,DEC] in rad
pos1 = [[1.42,1.620],[1.425,1.625]]
pos2 = [[1.52,1.620],[1.525,1.625]]
pos3 = [[1.42,1.820],[1.425,1.825]]


T = 1000.0                  #Period of observation[sec]
```

```python
num_pos1,t1, w_f1 = qua.get_num_photon(pos1,lam,s1,s2,T,'pos')
#num_neg1,t1, w_f1 = qua.get_num_photon(pos1,lam,s1,s2,T,'neg')

num_pos2,t2, w_f2 = qua.get_num_photon(pos2,lam,s1,s2,T,'pos')
#num_neg2,t2, w_f2 = qua.get_num_photon(pos2,lam,s1,s2,T,'neg')

num_pos3,t3, w_f3 = qua.get_num_photon(pos3,lam,s1,s2,T,'pos')
#num_neg3,t3, w_f3 = qua.get_num_photon(pos3,lam,s1,s2,T,'neg')
print(w_f1, w_f2, w_f3)
```

```python

plt.figure(figsize=(10,10))
plt.xlim([900.0,901.0])

plt.plot(t1[t1.size/2:],num_pos1[t1.size/2:],label='pos1, ref')
#plt.plot(t1.size[:],num_neg1[t1.size:],label='neg1, ref')

plt.plot(t2[t2.size/2:],num_pos2[t2.size/2:],label='pos2, D_RA')
#plt.plot(t2[:],num_neg2[:],label='neg2, D_RA')

plt.plot(t3[t3.size/2:],num_pos3[t3.size/2:],label='pos3, D_DEC')
#plt.plot(t3[:],num_neg3[t3.size/2:],label='neg3, D_DEC')

plt.xlabel("time [sec]");plt.ylabel("Number of Coincidence")
plt.legend() 
```
