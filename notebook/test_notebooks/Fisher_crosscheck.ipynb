```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.array([0.1508/0.09958294,0.085918/0.07041577,0.06617753/0.05749423,0.04679921/0.0445348,0.03007887/0.03149089])
rates = np.array([1,2,3,5,10])
fig = plt.figure(figsize=(7,4))
plt.rc('font', size=13)
ax = fig.add_subplot(111)
ax.set_xlabel('Pair Count Multiplier')
ax.set_ylabel(r'$\sigma_{mcmc}/\sigma_{fisher}$',size=16)
ax.scatter(rates,data)
plt.savefig('fisher_crosscheck', facecolor='white')
```
